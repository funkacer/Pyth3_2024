```python
import pandas as pd
```

```python
from collections import defaultdict

def getParams (f, count=10):
    linecount, wordcount = 0,0
    result = defaultdict(int)
    soubor = open(f, 'r')
    for line in soubor.readlines():
        #print("radek: {}".format(line.strip()))
        linecount += 1
        wordcount += len(line.strip().split())
        line = line.strip().split()
        for word in line:
            result[word]+=1
    soubor.close()
    top = {}
    #print("Soubor {} ma {} radku a {} slov".format(file, linecount, wordcount))
    #pojmenovany atribut key
    '''
    for word, freq in sorted(result.items(), key=lambda x:x[1], reverse=True)[:count]:
        top.update({word:freq})
    '''
    top = {word:freq for word, freq in sorted(result.items(), key=lambda x:x[1], reverse=True)[:count]}
    return linecount, wordcount, top
```

```python
files =[r'_data/test.txt',
        r'_data/anne.txt', r'_data/big.txt']
data = list(map(getParams,files))
```

```python
df = pd.DataFrame(data)
```

# objektove relacni mapovani

```python
from sqlalchemy import Column, Integer, String, create_engine
from sqlalchemy.ext.declarative import declarative_base as BASE
```

```python
conn = create_engine("mysql://root:root@127.0.0.1:8890/jiri")
```

```python
base = BASE()
base.metadata.create_all(conn)
```

```python
%sql desc jiri.gopas
```

```python
class Gopas(base):
    __tablename__ = "gopas"
    id = Column(Integer, primary_key = True)
    name = Column(String)
    salary = Column (Integer)
    def __str__(self):
        return "{}, {}, {}".format(self.id, self.name, self.salary)
```

```python
from sqlalchemy.orm import sessionmaker
```

```python
session = sessionmaker(bind=conn)
s=session()
```

```python
data = s.query(Gopas).all()
for record in data:
    #print(record.id, record.name, record.salary)
    print(record)
```

```python
newRec = Gopas(name="tibor", salary = 2400)
```

```python
s.add(newRec)
```

```python
s.commit()
```

```python
data = s.query(Gopas).all()
for record in data:
    print(record)
```
