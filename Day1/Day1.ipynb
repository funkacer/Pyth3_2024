# WebScraping

```python
wiki='https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India'
```

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
response = requests.get(wiki)
```

```python
response.__class__
```

```python
response.status_code
```

```python
soup=BeautifulSoup(response.content)
```

```python
soup.__class__
```

```python
soup.title
```

```python
soup.title.text
```

```python
soup
```

```python
atags=soup.findAll("a")
```

```python
atags
```

```python
for link in atags:
    print(link.get('href'))
```

```python
tables = soup.findAll("table")
```

```python
tables
```

```python
#wiki_tables = soup.find('table',{'class':'wikitable sortable plainrowheaders'})
wiki_tables = soup.find('table',{'class':'wikitable sortable'})
```

```python
wiki_tables
```

```python
allStates = {}
for row in wiki_tables.find_all('tr'):
    cells=row.findAll('td')
    states = row.findAll('th')
    if len(cells)==6:
        state = states[0].find(text=True).strip()
        capital = cells[1].find(text=True).strip()
        est = cells[4].find(text=True).strip()
        print("{} ({}) (established: {})".format(state, capital, est))
        allStates.update({state:{'capital':capital,"year":est}})
```

```python
allStates
```

```python
allStates["Goa"]
```

# Heureka.cz

```python
heur = "https://www.heureka.cz/?h%5Bfraze%5D=adidas+deodorant+woman"
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen
```

```python
#response = requests.get(heur)
html = urlopen(heur)
```

```python
#response.status_code
html.status
```

```python
heureka=BeautifulSoup(html.read())
```

```python
heureka
```

```python
allProds = heureka.find_all('a', class_='js-serpSpamScore', href = True)
```

```python
allProds.__class__
```

```python
first = allProds[0]
```

```python
count =0
for prod in allProds[::3]:
    print(prod.string, " ", count)
    count += 1
print ("Pocet produktu", count)
```

```python
popis = BeautifulSoup(requests.get(first['href']).content, 'html.parser')
```

```python
popis
```

```python
spec = popis.find_all(class_='product-desc')[0].get_text()
```
