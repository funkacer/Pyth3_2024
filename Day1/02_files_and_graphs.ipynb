```python
from collections import defaultdict

def getParams (f, count=10):
    linecount, wordcount = 0,0
    result = defaultdict(int)
    soubor = open(f, 'r')
    for line in soubor.readlines():
        #print("radek: {}".format(line.strip()))
        linecount += 1
        wordcount += len(line.strip().split())
        line = line.strip().split()
        for word in line:
            result[word]+=1
    soubor.close()
    top = {}
    #print("Soubor {} ma {} radku a {} slov".format(file, linecount, wordcount))
    #pojmenovany atribut key
    '''
    for word, freq in sorted(result.items(), key=lambda x:x[1], reverse=True)[:count]:
        top.update({word:freq})
    '''
    top = {word:freq for word, freq in sorted(result.items(), key=lambda x:x[1], reverse=True)[:count]}
    return linecount, wordcount, top
    
```

```python
import matplotlib.pyplot as plt
```

```python
x = {"Ahoj":2, "Ne":3, "Ano":5}
```

```python
def getGraph(nazev, di, angle = 0):
    plt.bar(range(len(di)),di.values())
    plt.xticks(range(len(di)), di.keys(), rotation=angle)
    plt.title(nazev)
    plt.show()
```

```python
getGraph("Dict",x)
```

```python
for file, params in result.items():
    freq=params[-1]
    getGraph(file,freq)
```

```python
def getPie(nazev, x):
    fig = plt.figure(1, figsize=(6,6),dpi=100)
    az = fig.add_axes([0.6,0.1,0.6,0.6])
    key=list(x.keys())
    value = list(x.values())
    expl = [0.0 for _ in range(len(x))]
    expl[0] = 0.15
    az.pie(value,labels = key, explode = expl, shadow = True, autopct="%1.2f%%")
    plt.title(nazev)
    plt.show()
```

```python
getPie("AAA", x)
```

```python
for file, params in result.items():
    freq=params[-1]
    getPie(file, freq)
```
